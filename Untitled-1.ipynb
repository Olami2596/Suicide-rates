## Suicide Rate Analysis 

```python
#importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#making a copy of the dataset before any wrangling procedure is carried out
df_copy = df.copy()
df_copy
```

Wrangling Process

```python
#drop the Unnamed 7 and Unnamed : 8 column because they are not needed
df_copy = df_copy.drop(['Unnamed: 7', 'Unnamed: 8'], axis = 1)
df_copy
```

```python
#drop nan values
df_copy = df_copy.dropna()
```

```python
#when trying to change columns to int some rows contained Null so they had to be dropped
remove_null = (df_copy == "Null").any(axis=1)
df_copy = df_copy[~remove_null]
```

```python
#when trying to change columns to int some rows contained unknown so they had to be dropped
remove_unknown = (df_copy == "Unknown").any(axis=1)
df_copy = df_copy[~remove_unknown]
```

```python
#creating a function to remove the comma's in the gdp column and convert the column into int 
def comma_remover(x):
    return int(x.replace(',', ''))
```

```python
df_copy.iloc[:,6] = df_copy.iloc[:,6].apply(comma_remover)
```

```python
df_copy.iloc[:,6]
```

```python
#change the suicides_no column from object to int
df_copy['suicides_no'] = pd.to_numeric(df_copy['suicides_no'])
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
#fixing the age column
df_copy['age'] = df_copy['age'].str.strip(' years')
df_copy['age'] = df_copy['age'].str.strip(' +')
```

```python
df_copy['start_age'], df_copy['end_age'] = df_copy['age'].str.split('-').str
```

```python
df_copy = df_copy.dropna()
```

```python
df_copy['start_age'] = df_copy.loc[:,'start_age'].astype(int)
df_copy['end_age'] = df_copy.loc[:,'end_age'].astype(int)
```

```python
df_copy['mean_age'] = (df_copy['start_age'] + df_copy['end_age']) / 2
```

```python
df_copy = df_copy.drop(['start_age', 'end_age','age'], axis = 1)
```

```python
df_copy
```

Question


Add a new column “generation” and fill up its data according to below criteria


Value Criteria


Lost Generation born between 1883 - 1900


G.I. Generation born between 1901 - 1927


Silent born between 1928 - 1945


Boomers born between 1946 – 1964


Generation X born between 1965 - 1980


Millennials born between 1981 - 1996


Generation Z born between 1997 – 2012


Generation A born between 2013 – 2025 

```python
def generation(year):
    if 1883 <= year <= 1900:
        return 'Lost Generation'
    elif 1901 <= year <= 1927:
        return 'G.I. Generation'
    elif 1928 <= year <= 1945:
        return 'Silent'
    elif 1946 <= year <= 1964:
        return 'Boomers'
    elif 1965 <= year <= 1980:
        return 'Generation X'
    elif 1981 <= year <= 1996:
        return 'Millenials'
    elif 1997 <= year <= 2012:
        return 'Generation Z'
    elif 2013 <= year <= 2025:
        return 'Generation A'
```

```python
df_copy['generation'] = df['year'].apply(generation)
```

```python
df_copy
```

Add a new column “gdp_per_capita” and fill its data. GDP per Capita of a country is GDP divided by population of that country.

Question


Use appropriate visual notation to visualise total suicides over years. Describe your findings
